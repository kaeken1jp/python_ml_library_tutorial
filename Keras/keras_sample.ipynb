```python
#kerasのミニマムなサンプル
#https://qiita.com/Ishotihadus/items/c2f864c0cde3d17b7efb
#適当なn次元の配列を入力したら、その和が特定の値より大きいときに1と教えてくれるニューラルネットを作る

import numpy as np
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Activation

#5次元のデータを250個作成
data = np.random.rand(250,5)
#サンプル表示
data_sample = np.random.rand(2,5)
print('====')
print(data_sample)

#250個の各データに対して5次元の和をとり、それが2.5より大きいときに1となる（そうでないときは0となる）ような配列
#https://keras.io/ja/utils/np_utils/
labels = np_utils.to_categorical((np.sum(data, axis=1) > 2.5) * 1)
labels_sample = np_utils.to_categorical((np.sum(data_sample, axis=1) > 2.5) * 1)
print('====')
print(np.sum(data_sample, axis=1))
print('====')
print((np.sum(data_sample, axis=1) > 2.5) * 1)
print('====')
print(labels_sample)




```

```python
#Sequentialはただ層を積み上げるだけの単純なモデル
model = Sequential([Dense(20, input_dim=5), Activation('relu'), Dense(2, activation='softmax')])
#コンパイル
model.compile('rmsprop', 'categorical_crossentropy', metrics=['accuracy'])
#学習
model.fit(data, labels, nb_epoch=300, validation_split=0.2)

test = np.random.rand(200, 5)
#予測
predict = np.argmax(model.predict(test), axis=1)
real = (np.sum(test, axis=1) > 2.5) * 1
print(sum(predict == real) / 200.0)
```
